```python
import numpy as np
x = np.array([1,2,3])
y = np.array([4,5,6])
sum = np.add(x,y)
print(sum)
```

```python
print(np.log(sum))
```

```python
arr_power = np.power(sum, 2)
print(arr_power)
print(np.exp(x))
```

```python
#comparison and logical UFuncs
arr_2d = np.array([[2,6,8],
              [10,12,14],
              [16,18,20]])
cond1 = arr_2d >8
cond2 = arr_2d <15
print(cond1)
print("\n", cond2)
print("\n", np.logical_and(cond1,cond2))
```

```python
#Broadcasting
arr_b= np.array([2,4,6])
print(arr_b + 10)
```

```python
#Linear algebra
A = np.array([[2,4],
              [6,8]])
B = np.array([[3,6],
              [9,12]])
print("Matrix Multiplication \n", A @ B)
print("\nInverse of A", np.linalg.inv(A))
```

```python
#Numpy Module
rng = np.random.default_rng()
rand_int = rng.integers(low=0, high=10, size=5)
print(rand_int)
print("\nRandom float", rng.random(5))
```

```python
#Normally distributed random numbers
n_ran = rng.normal(loc=0, scale=1, size=5)
print(n_ran)
```

```python
#Shuffle and permutation
array_1 = np.array([10,20,30,50])
shuffled_arr = rng.shuffle(array_1)
print(shuffled_arr)
print(rng.permutation(array_1))
```
