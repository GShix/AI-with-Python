```python
#Creating DataFrame
import pandas as pd
import numpy as np
import random

random.seed(101)
#defining realistic names
names = ['Dambar', 'Nischal', 'Tarak', 'Buddhi']
#genders corresponding to names
genders = ['Male', 'Male', 'Male', 'Male']

#generate random data
ages = [random.randint(21,25) for _ in range(4)]
marks = [random.randint(60,95) for _ in range(4)]
subjects = random.choices(['Python', 'Statistics', 'Generative AI'], k=4)

#create DataFrame
df = pd.DataFrame({
    'Name': names,
    'Gender': genders,
    'Age': ages,
    'Marks': marks,
    'Subjects': subjects
})

df
```
