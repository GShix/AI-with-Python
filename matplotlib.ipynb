```python
#Matplotlib
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2*np.pi,100)
y = np.sin(x)

plt.figure(figsize=(6,4))
plt.plot(x,y, label='Sin(x)', color='limegreen', linestyle='-', linewidth=2)
plt.title("Line plot of Sine wave\n", fontsize=15)
plt.xlabel("X values", fontsize=12)
plt.ylabel("Y values", fontsize=12)
plt.legend(loc=3)
plt.grid(True)
plt.show()
```
