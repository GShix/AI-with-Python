```python
#Matplotlib
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 2*np.pi,100)
y = np.sin(x)

plt.figure(figsize=(6,4))
plt.plot(x,y, label='Sin(x)', color='limegreen', linestyle='-', linewidth=2)
plt.title("Line plot of Sine wave\n", fontsize=15)
plt.xlabel("X values", fontsize=12)
plt.ylabel("Y values", fontsize=12)
plt.legend(loc=3)
plt.grid(True)
plt.show()
```

```python
#Scatter plot
np.random.seed(0)
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
plt.figure(figsize=(6,4))
plt.scatter(x_scatter, y_scatter, c='red', marker='.', alpha=0.8)
plt.title("Scatter Plot")
plt.xlabel("x Values")
plt.ylabel("Y Values")
plt.grid(True)
plt.show()
```

```python
#Bar graph: Categorical data
candidates = ['Balen Shah', 'KP Oli']
votes = [68348, 18734]
plt.figure(figsize=(6,4))
plt.bar(candidates, votes, color='deepskyblue')
plt.title("Bar Graph - Balen vs KP Oli")
plt.xlabel("Candidates")
plt.ylabel("Votes")
plt.show()
```

```python
#Histogram - continuous data
data = np.random.randn(1000)
plt.figure(figsize=(6,4))
plt.hist(data, bins=30, color='red', alpha=0.7, edgecolor='black')
plt.title("Histogram")
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.show()
```
